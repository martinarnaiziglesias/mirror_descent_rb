```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.patches import Patch

# Assuming 'm' is defined and the np.load functions have already been run for both SGD and SMD

# SGD Data
df_sgd = pd.DataFrame(columns=[('SGD tamed_1', 10),
                      ('SGD tamed_1',25),
                      ('SGD tamed_1', 50),
                      ('SGD tamed_1', 100),
                      ('SGD tamed_1', 250),
                      ("SGD tamed_2",10),
                      ("SGD tamed_2", 25),
                      ("SGD tamed_2", 50),
                        ("SGD tamed_2",100),
                        ("SGD tamed_2", 250),
                        ('SGD tamed_3', 10),
                        ('SGD tamed_3', 25),
                        ('SGD tamed_3', 50),
                        ('SGD tamed_3', 100),
                        ('SGD tamed_3', 250)])

df_sgd[("SGD tamed_1", 10)] = np.load('10_sgd_tamed_norms.npy')[:m, 2]
df_sgd[("SGD tamed_1", 25)] = np.load('25_sgd_tamed_norms.npy')[:m, 2]
df_sgd[("SGD tamed_1", 50)] = np.load('50_sgd_tamed_norms.npy')[:m, 2]
df_sgd[("SGD tamed_1", 100)] = np.load('100_sgd_tamed_norms.npy')[:m, 2]
df_sgd[("SGD tamed_1", 250)] = np.load('250_sgd_tamed_norms.npy')[:m, 2]
df_sgd[("SGD tamed_2", 10)] = np.load('10_sgd_tamed_norms.npy')[:m, 5]
df_sgd[("SGD tamed_2", 25)] = np.load('25_sgd_tamed_norms.npy')[:m, 5]
df_sgd[("SGD tamed_2", 50)] = np.load('50_sgd_tamed_norms.npy')[:m, 5]
df_sgd[("SGD tamed_2", 100)] = np.load('100_sgd_tamed_norms.npy')[:m, 5]
df_sgd[("SGD tamed_2", 250)] = np.load('250_sgd_tamed_norms.npy')[:m, 5]
df_sgd[("SGD tamed_3", 10)] = np.load('10_sgd_tamed_norms.npy')[:m, 8]
df_sgd[("SGD tamed_3", 25)] = np.load('25_sgd_tamed_norms.npy')[:m, 8]
df_sgd[("SGD tamed_3", 50)] = np.load('50_sgd_tamed_norms.npy')[:m, 8]
df_sgd[("SGD tamed_3", 100)] = np.load('100_sgd_tamed_norms.npy')[:m, 8]
df_sgd[("SGD tamed_3", 250)] = np.load('250_sgd_tamed_norms.npy')[:m, 8]

df_sgd.columns = pd.MultiIndex.from_tuples(df_sgd.columns)

# SMD Data
df_smd = pd.DataFrame(columns=[('SMD_1', 10),
                      ('SMD_1',25),
                      ('SMD_1', 50),
                      ('SMD_1', 100),
                      ('SMD_1', 250),
                      ("SMD_2",10),
                      ("SMD_2", 25),
                      ("SMD_2", 50),
                        ("SMD_2",100),
                        ("SMD_2", 250),
                        ('SMD_3', 10),
                        ('SMD_3', 25),
                        ('SMD_3', 50),
                        ('SMD_3', 100),
                        ('SMD_3', 250)])

df_smd[("SMD_1", 10)] = np.load('10_smd_norms.npy')[:m, 2]
df_smd[("SMD_1", 25)] = np.load('25_smd_norms.npy')[:m, 2]
df_smd[("SMD_1", 50)] = np.load('50_smd_norms.npy')[:m, 2]
df_smd[("SMD_1", 100)] = np.load('100_smd_norms.npy')[:m, 2]
df_smd[("SMD_1", 250)] = np.load('250_smd_norms.npy')[:m, 2]
df_smd[("SMD_2", 10)] = np.load('10_smd_norms.npy')[:m, 5]
df_smd[("SMD_2", 25)] = np.load('25_smd_norms.npy')[:m, 5]
df_smd[("SMD_2", 50)] = np.load('50_smd_norms.npy')[:m, 5]
df_smd[("SMD_2", 100)] = np.load('100_smd_norms.npy')[:m, 5]
df_smd[("SMD_2", 250)] = np.load('250_smd_norms.npy')[:m, 5]
df_smd[("SMD_3", 10)] = np.load('10_smd_norms.npy')[:m, 8]
df_smd[("SMD_3", 25)] = np.load('25_smd_norms.npy')[:m, 8]
df_smd[("SMD_3", 50)] = np.load('50_smd_norms.npy')[:m, 8]
df_smd[("SMD_3", 100)] = np.load('100_smd_norms.npy')[:m, 8]
df_smd[("SMD_3", 250)] = np.load('250_smd_norms.npy')[:m, 8]

df_smd.columns = pd.MultiIndex.from_tuples(df_smd.columns)

# Melt and combine the datasets
df_sgd_melted = pd.melt(df_sgd.stack().reset_index().drop(columns='level_0'),
                        id_vars=['level_1'], 
                        value_vars=['SGD tamed_1', 'SGD tamed_2', 'SGD tamed_3'], 
                        var_name='method', 
                        value_name='value')

df_smd_melted = pd.melt(df_smd.stack().reset_index().drop(columns='level_0'),
                        id_vars=['level_1'], 
                        value_vars=['SMD_1', 'SMD_2', 'SMD_3'], 
                        var_name='method', 
                        value_name='value')

# Add an identifier to the method for SGD or SMD
df_sgd_melted['Algorithm'] = 't-SGD'
df_smd_melted['Algorithm'] = 'SMD'

# Combine both melted dataframes
# combined_df = pd.concat([df_sgd_melted, df_smd_melted])
combined_df = pd.concat([df_smd_melted, df_sgd_melted,])

palette = {
    'SGD tamed_1': 'Blue',
    'SGD tamed_2': 'MediumBlue',
    'SGD tamed_3': 'DarkBlue',
    'SMD_1': 'ForestGreen',
    'SMD_2': 'Green',
    'SMD_3': 'DarkGreen'
}

# Hatching patterns for _1, _2, _3
hatches = ['O']*5 + ['++']*5 + ['.']*5 + ['O']*5 + ['++']*5 + ['.']*5 + ['']*6

# Plot the combined grouped boxplot
plt.figure(figsize=(2*6.4, 4.8))
box_plot = sns.boxplot(x='level_1', y='value', data=combined_df, hue='method', showfliers=True, width=.75, fliersize=1, palette=palette)
plt.yscale('log')
# plt.title('Error Comparison: SGD vs SMD')

# Add hatches to distinguish the _1, _2, and _3 categories
for i, patch in enumerate(box_plot.patches):
    hatch = hatches[i]  # Cycle through hatches
    patch.set_hatch(hatch)

# Add labels and legends
plt.ylabel(r'$|| u^k -  u^*||_1$')
plt.xlabel('Number of assets')

# Legend for the colors (SGD vs SMD)
legend_elements_1 = [Line2D([0], [0], color='Green', lw=4, label='SMD'),
                     Line2D([0], [0], color='MediumBlue', lw=4, label='t-SGD')
                     ]

# Legend for the hatches (_1, _2, _3)
legend_elements_2 = [Patch(facecolor='white', edgecolor='black', hatch=hatches[0], label=r'$k=3\times10^5$'),
                     Patch(facecolor='white', edgecolor='black', hatch=hatches[5], label=r'$k=6\times10^5$'),
                     Patch(facecolor='white', edgecolor='black', hatch=hatches[10], label=r'$k=9\times10^5$')]

# First legend: Color legend for SGD and SMD


# Second legend: Hatching legend for _1, _2, _3
plt.gca().add_artist(plt.legend(handles=legend_elements_2, title="Iterations", bbox_to_anchor=(0.22, 0.275), fontsize=9, title_fontsize=10))
plt.gca().add_artist(plt.legend(handles=legend_elements_1, loc='lower left', title="Algorithm", fontsize=9, title_fontsize=10))
plt.savefig('error_comparison_portfolio.pdf')
plt.show()
plt.tight_layout()
```

```python
df_1 = pd.DataFrame(columns=[('t-SGD', 'n_1'), ('t-SGD', 'n_2'), ('t-SGD', 'n_3')], index=[10, 25, 50, 100, 250])
df_1.loc[:,:] = df_sgd.median().unstack().T.values

df_2 = pd.DataFrame(columns=[('SMD', 'n_1'), ('SMD', 'n_2'), ('SMD', 'n_3')], index=[10, 25, 50, 100, 250])
df_2.loc[:,:] = df_smd.median().unstack().T.values

df_mean = pd.concat([df_1,df_2], axis=1)
df_mean.columns = pd.MultiIndex.from_tuples(df_mean.columns)
df_mean.columns = df_mean.columns.swaplevel(0, 1)
df_mean = df_mean.sort_index(axis=1, level=0)

from scipy import stats

df_1 = pd.DataFrame(columns=[('t-SGD', 'n_1'), ('t-SGD', 'n_2'), ('t-SGD', 'n_3')], index=[10, 25, 50, 100, 250])
df_1.loc[:,:] = pd.DataFrame(stats.median_abs_deviation(df_sgd), index=df_sgd.columns)[0].unstack().T.values

df_2 = pd.DataFrame(columns=[('SMD', 'n_1'), ('SMD', 'n_2'), ('SMD', 'n_3')], index=[10, 25, 50, 100, 250])
df_2.loc[:,:] = pd.DataFrame(stats.median_abs_deviation(df_smd), index=df_smd.columns)[0].unstack().T.values

df_mad = pd.concat([df_1,df_2], axis=1)
df_mad.columns = pd.MultiIndex.from_tuples(df_mad.columns)
df_mad.columns = df_mad.columns.swaplevel(0, 1)
df_mad = df_mad.sort_index(axis=1, level=0)

df_output = pd.DataFrame(columns=df_mean.columns, index=df_mean.index)
for i in df_mean.columns:
    for j in df_mean.index:
        df_output.loc[j,i] = f"{df_mean.loc[j,i]*10000:.2f} ({df_mad.loc[j,i]*10000:.2f})"
print(df_output.to_latex())

```

```python
from scipy import stats

df_final = pd.DataFrame(columns=[('MAD', 'SMD'), ('MAD', 't-SGD'),('Volatility', 'SMD'), ('Volatility', 't-SGD'),('Variantile', 'SMD'), ('Variantile', 't-SGD')],
                        index=[10, 25, 50, 100, 250])

c = 10000
for d in [10, 25, 50, 100, 250]:
    arr = np.load(str(d)+'_smd_norms_volatility.npy')
    df_final[('Volatility', 'SMD')].loc[d] = f"{np.median(arr)*c:.2f} ({stats.median_abs_deviation(arr)*c:.2f})"

    arr = np.load(str(d)+'_sgd_norms_volatility.npy')
    df_final[('Volatility', 't-SGD')].loc[d] = f"{np.median(arr)*c:.2f} ({stats.median_abs_deviation(arr)*c:.2f})"

    arr = np.load(str(d)+'_smd_norms_mad.npy')
    df_final[('MAD', 'SMD')].loc[d] = f"{np.median(arr)*c:.2f} ({stats.median_abs_deviation(arr)*c:.2f})"

    arr = np.load(str(d)+'_sgd_norms_mad.npy')
    df_final[('MAD', 't-SGD')].loc[d] = f"{np.median(arr)*c:.2f} ({stats.median_abs_deviation(arr)*c:.2f})"

    arr = np.load(str(d)+'_smd_norms_variantile.npy')
    df_final[('Variantile', 'SMD')].loc[d] = f"{np.median(arr)*c:.2f} ({stats.median_abs_deviation(arr)*c:.2f})"

    arr = np.load(str(d)+'_sgd_norms_variantile.npy')
    df_final[('Variantile', 't-SGD')].loc[d] = f"{np.median(arr)*c:.2f} ({stats.median_abs_deviation(arr)*c:.2f})"
```

```python
sgd_errors = np.load('sgd_errors_tamed.npy')
smd_errors = np.load('smd_errors_tamed.npy')
```

```python
nan_index = []
idx = 0
for i in sgd_errors:
    if True in np.isnan(i):
        nan_index.append(idx)
    idx+=1
```

```python
nan_index
```

```python
sgd_errors = np.delete(sgd_errors, nan_index, axis=0)
smd_errors = np.delete(smd_errors, nan_index, axis=0)
```

```python
n_val=1000000
freq_error = 50
ns = []
for k in range(1, n_val):
    if k%freq_error==0:
        ns.append(k)
```

```python
fig, ax = plt.subplots(figsize=(6,4))

ax.plot(ns, np.quantile(smd_errors, q=.5, axis=0), label='SMD', color='blue', lw=.5)
ax.fill_between(ns, np.quantile(smd_errors, q=.025, axis=0), np.quantile(smd_errors, q=.975, axis=0), color='blue', alpha=0.2)

ax.plot(ns, np.quantile(sgd_errors, q=.5, axis=0), label='t-SGD', color='green', lw=.5)
ax.fill_between(ns, np.quantile(sgd_errors, q=.025, axis=0), np.quantile(sgd_errors, q=.975, axis=0), color='green', alpha=0.2)

ax.set_yscale('log')
# ax.set_xscale('log')
ax.legend()
ax.set_ylabel(r'$\Gamma\left(y^k\right) - \Gamma\left(y^*\right)$')
ax.set_xlabel('Iterations')

# plt.savefig('error_iterations_toy_model.pdf')
# plt.show()
# plt.tight_layout()
```
